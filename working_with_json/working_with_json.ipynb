```python
import json

# opening the states.json file
with open('states.json') as f:
    data = json.load(f)

for state in data['states']:
    del state['area_codes']

#now saving the python dictionary into json file
with open('new_states.json', 'w') as f:
    json.dump(data, f, indent=2)
```

```python
# we got a yahoo api that convert USD to other currency

from urllib.request import urlopen

with urlopen("https://finance.yahoo.com/webservice/v1/symbols/allcurrencies/quote?format=json") as response:
    source = response.read()
    
data = json.loads(source.decode("utf-8"))

print(json.dumps(data, indent=2))
```

```python
usd_rate = dict()

for item in data['list']['resources']:
    name = item['resource']['fields']['name']
    price = item['resource']['fields']['price']
    
    usd_rate[name] = price
    
input_name = input("enter the input conversion name : ")

num_of_dollars = int(input("enter the number of dollars : "))

print(num_of_dollars*float(usd_rate[input_name]))
```
